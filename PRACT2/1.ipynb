```python
import logging

logging.basicConfig(
    filename="secondLargest.log",
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s"
)

def second_largest(nums: list) -> int:
    """
    Find second largest element in a list.
    Parameters:
        nums (list): List of integers
    Returns:
        int: Second largest element
    """
    logging.info("Finding second largest element")

    largest = -1
    secondLargest = -1

    for num in nums:
        if num > largest:
            secondLargest= largest
            largest = num
        elif num > secondLargest and num != largest:
            secondLargest = num

    logging.info(f"Largest: {largest}, Second Largest: {secondLargest}")

numbers = [10, 20, 4, 45, 99]
result = second_largest(numbers)
```

```python

def is_sorted(nums: list) -> bool:
    logging.info("Checking if list is sorted in ascending order")

    for i in range(len(nums) - 1):
        if nums[i] > nums[i + 1]:
            logging.info("List is NOT sorted")
            return False

    logging.info("List is sorted")
numbers = [1,2,30,4,5]
result = is_sorted(numbers)
```

```python
def reverseStr(S:str)->str:
    logging.info("Reversing the string")
    rev=""
    for i in range(len(S)-1,-1,-1):
        rev+=S[i]
    logging.info(f"Reversed string: {rev}")

reverseStr("Hello World")
```

```python
def removeDuplicates(nums: list) -> list:
    logging.info("Removing duplicates while preserving order")
    uniqueList = []
    for num in nums:
        if num not in uniqueList:
            uniqueList.append(num)
    logging.info(f"List after removing duplicates: {uniqueList}")
numbers = [1,2,2,3,4,3,5,1]
result = removeDuplicates(numbers)
```

```python
def findMissing(nums:list)->int:
    """Finding the missing number in a list of integers from 1 to n.
    parameters it takes a list of integers as input 
    returns the intger value
    """
    logging.info("Finding the missing number in the list")
    n=len(nums)+1
    sum=n*(n+1)//2
    actualSum=0
    for i in nums:
        actualSum+=i
    ans=sum-actualSum
    logging.info(f"The missing VAlue is {ans}")
nums=[1,2,3,5,6]
findMissing(nums)
```
