```python
with open('a.txt','w') as f:
    f.write(''' jfhxkjbgk\
            nfjdsndxgnjbk\
            ksdnfjl;f\
            dksjfla''')
    
#with open print genreally on the same line but we can use end to change the default behaviour of print function WHERE AS without with open print will print on the next line.
```

```python
with open('data.txt','r') as f:
    word=(f.read().split())
unique_word=set(word)
print(unique_word)
```

```python
with open('data.txt','r') as f:
    word=(f.read().split())
    print(word)
    #print(len(word))
    freq={}
    for i in word:
        if i in freq:
            freq[i]+=1
        else:
            freq[i]=1
print(freq)

```

```python
import  csv
with open('data.csv','w',newline='') as file:
    writer=csv.writer(file)
    writer.writerow(['Name','Age','Course'])
    writer.writerow(['Devesh',20,'Python'])
```

```python
import csv
rows=[['Name','Age','Course'],['Devesh',20,'Python'],['John',22,'Java'],['Alice',19,'Cloud Computing']]
with open('data.csv','w',newline='') as file:
    
    writer=csv.writer(file)
    writer.writerows(rows)
```

```python
import csv
with open('data.csv','w',newline='') as file:
    writer=csv.DictWriter(file,fieldnames=['Name','Age','Course'])
    writer.writeheader()
    writer.writerow({'Name':'Devesh','Age':20,'Course':'Python'})
    writer.writerow({'Name':'John','Age':22,'Course':'Java'})
```

```python
with open("data.csv","r") as file:
    reader=csv.DictReader(file)
    for row in reader:
        print(row)
```
