### Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.random.randint(1,50,10)
y=np.random.randint(1,50,10)

print (x,y,sep='\n')
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,color='red',linestyle='dotted')
plt.show()
```

```python
plt.title("Random chart")
plt.xlabel('x_axis label')
plt.ylabel('Y_axis label')
plt.plot(x,y,color='red',linestyle='dotted')
plt.show()
```

```python
plt.title("Random chart")
plt.xlabel('x_axis label')
plt.ylabel('Y_axis label')
plt.xticks(x)
plt.yticks(y)
plt.plot(x,y,color='red',linestyle='dotted')
plt.show()
```

### LINE STYLE

-  ->solid(default)
--  -> dashed
-.  ->dsh-dot
:   ->dotted

```python
plt.title("Random chart")
plt.xlabel('x_axis label')
plt.ylabel('Y_axis label')
plt.xticks(x)
plt.yticks(y)
plt.plot(x,y,color='red',linestyle='-.')
plt.show()
```

```python
plt.title("Random chart")
plt.xlabel('x_axis label')
plt.ylabel('Y_axis label')
plt.xticks(x)
plt.yticks(y)
plt.plot(x,y,color='red',linestyle='-',alpha=0.3)  #ALPHA USED FOR INTENSITY
plt.show()
```

```python
x=np.arange(1,11)
y=np.sin(x)
z=np.cos(x)

print(x,y,z ,sep='\n\n')
```

```python
plt.plot(x,y,label='SINQQ')
plt.plot(x,z,label='COSQ')
plt.title("SINQ & COSQ")
plt.xlabel("Data points")
plt.ylabel("Associate values")
plt.legend()
plt.show()
```

```python

plt.figure(figsize=(8,3))
plt.plot(x,y,label='SINQQ')
plt.plot(x,z,label='COSQ')
plt.title("SINQ & COSQ")
plt.xlabel("Data points")
plt.ylabel("Associate values")
plt.legend()
plt.show()
```

```python

plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.plot(x,y,label='SINQQ')
plt.plot(x,z,label='COSQ')
plt.title("SINQ & COSQ")
plt.xlabel("Data points")
plt.ylabel("Associate values")
plt.legend()

plt.subplot(1,2,2)
plt.plot(x,y,label='SINQQ',color='green')
plt.plot(x,z,label='COSQ')
plt.title("SINQ & COSQ")
plt.xlabel("Data points")
plt.ylabel("Associate values")


plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.subplot(3,3,1)
plt.plot(x,y,label='SINQQ')
plt.plot(x,z,label='COSQ')
plt.title("SINQ & COSQ")
plt.xlabel("Data points")
plt.ylabel("Associate values")
plt.legend()
plt.grid()


plt.subplot(3,3,5)
plt.plot(x,y,label='SINQQ',color='green')
plt.plot(x,z,label='COSQ')
plt.title("SINQ & COSQ")
plt.xlabel("Data points")
plt.ylabel("Associate values")
plt.legend(loc='lower right',fontsize=7)
plt.grid()


plt.subplot(3,3,9)
plt.plot(x,y,label='SINQQ',color='yellow')
plt.plot(x,z,label='COSQ')
plt.title("SINQ & COSQ")
plt.xlabel("Data points")
plt.ylabel("Associate values")
plt.legend()
plt.grid()
plt.show()
```
