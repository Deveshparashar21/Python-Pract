```python
class Ten:
    name="Devesh"
    loc="Munger"
    marks=75
    def __init__(self,name,loc,marks):
        self.name=name
        self.loc=loc
        self.marks=marks

    def display(self):
        print("Name:",self.name)
        print("Location:",self.loc)
        print("Marks:",self.marks)
    
class Twelve(Ten):
    def __init__(self,name,loc,marks,stream):
        super().__init__(name,loc,marks)
        self.stream=stream
    def display(self):
         super().display()
         print("Stream:",self.stream)

class Bechlor(Twelve):
    def __init__(self,name,loc,marks,stream,Branch,Uni):
        super().__init__(name,loc,marks,stream)
        self.Branch=Branch
        self.Uni=Uni
    def display(self):
        super().display()
        print("Branch:",self.Branch)
        print("Uni:",self.Uni)
    

    
s1=Bechlor("Devesh","Munger",75,"Science","CSE","HIT")
s1.display()

    
```

```python
#Hybrid inheritance on school managment system
class Person:
    def __init__(self,name,age):
        self.name=name
        self.age=age
class Student(Person):
    def __init__(self,name,age,rollno):
        super().__init__(name,age)
        self.rollno=rollno
class Teacher(Person):
    def __init__(self,name,age,subject):
        super().__init__(name,age)
        self.subject=subject
class Academics:
    def __init__(self,course=None,CGPA=None):
        self.course=course
        self.CGPA=CGPA
class Sports:
    def __init__(self,sport_name=None,level=None):
        self.sport_name=sport_name
        self.level=level
#Hybrid class
class Allronder(Student,Academics,Sports):
    def __init__(self,name,age,rollno,is_academic=False,is_sport=False,course=None,CGPA=None,sport_name=None,level=None):
        super().__init__(name,age,rollno)
        self.is_academic=is_academic
        self.is_sport=is_sport

        if self.is_academic:
            Academics.__init__(self,course,CGPA)
        if self.is_sport:
            Sports.__init__(self,sport_name,level)
    def display(self):
        print("Name:",self.name)
        print("Age:",self.age)
        print("Roll No:",self.rollno)
                                                                                                
        if self.is_academic:
            print("Course:",self.course)
            print("CGPA:",self.CGPA)
        if self.is_sport:
            print("Sport Name:",self.sport_name)
            print("Level:",self.level)
```

```python
#Academic student
s1=Allronder("Devesh",20,101,is_academic=True,course="B.Tech",CGPA=8.5)
s1.display()
print("\n")
s2=Allronder("Rahul",21,102,is_academic=True,is_sport=True,course="B.Tech",CGPA=7.82,sport_name="Football",level="State")
s2.display()
```
