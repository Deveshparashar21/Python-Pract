```python
nums = [1, 2, 3, 1, 2]
if nums[:2] == nums[-2:]:
    print("Valid slice")
else:
    print("Invalid slice")
```

```python
s = "RSr"
if s == s[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
s = "code9"
lastchar = s[-1]
if lastchar.isdigit():
    print("Ends with digit")
elif lastchar.isalpha():
    print("Ends with character")
else:
    print("Ends with special symbol")
```

```python
s = "Devesh"
if s[0].isupper():
    print("Starts with uppercase")
elif s[-1].islower():
    print("Ends with lowercase")
else:
    print("Other case")
```

```python
nums = [4, 3, 1, 2]
mid = len(nums) // 2
firsthalf = sum(nums[:mid])
secondhalf = sum(nums[mid:])
if firsthalf > secondhalf:
    print("First half greater")
elif firsthalf == secondhalf:
    print("Equal halves")
else:
    print("Second half greater")
```

```python
s = "hello"
if len(s) >= 5:
    mid = len(s) // 2
    middle_char = s[mid]

    if middle_char.lower() in "aeiou":
        print("Middle character is a vowel")
    else:
        print("Middle character is not a vowel")
else:
    print("String is less than 5 Char")
```

```python
s = "ABZ"
if s.startswith("A"):
    if s.endswith("Z"):
        print("Valid AZ string")
    else:
        print("Starts with A but invalid end")
else:
    print("Invalid string")
```

```python
s = "12ab"
if len(s) >= 4 and s[:2].isdigit():
    if s[-2:].isalpha():
        print("Valid format")
    else:
        print("Invalid format")
else:
    print("Invalid format")
```

```python
nums = [1, 2, 3, 1, 2, 3]
if len(nums) % 2 == 0:
    if nums[:2] == nums[-2:]:
        print("Slice with same elements")
    else:
        print("Slice with different elements")
else:
    print("Odd length list")
```
