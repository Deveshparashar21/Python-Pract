```python
from abc import ABC,abstractmethod
class BankAccount:
    def __init__(self,customer_name,DOB):
        self.customer_name=customer_name
        self.DOB=DOB
    
    @abstractmethod
    def deposit(self):
        pass

    @abstractmethod
    def withdraw(self):
        pass
            
class SavingAccount(BankAccount):
    def __init__(self,customer_name,DOB,min_balance):
        super().__init__(customer_name,DOB)
        self.min_balance=min_balance
        self.curr_bal=self.min_balance
    def deposit(self,amount):
       if amount>0:
           self.curr_bal+=amount
    def withdraw(self,amount):
        if self.curr_bal-amount>=self.min_balance:
            self.curr_bal-=amount

class currentAccount(BankAccount):
    def __init__(self,customer_name,DOB,overdraft_limit):
        super().__init__(customer_name,DOB)
        self.overdraft_limit=overdraft_limit
        self.curr_bal=overdraft_limit
    def deposit(self,amount):
       if amount>0:
           self.curr_bal+=amount
           print(f"Current balance after deposit is {self.curr_bal} ")
    def withdraw(self,amount):
        if self.curr_bal-amount>=-self.overdraft_limit:
            self.curr_bal-=amount
        if self.curr_bal<self.overdraft_limit:
            print(f"overdraft limit left {self.curr_bal}")
            print(f"Overdraft limit exceeded.add {self.overdraft_limit-self.curr_bal} till next month to make it stable")
obj2=currentAccount("Devesh","01-01-2000",5000)
obj2.deposit(2000)
obj2.withdraw(6000)
           
           
```
