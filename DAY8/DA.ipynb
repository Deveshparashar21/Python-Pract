```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv('List of Countries by Sugarcane Production.csv')
df.head()
```

```python
df.drop(columns='Unnamed: 0' , inplace=True)
df.head()
```

```python
df.info()
```

```python
df['Production (Tons)']=df['Production (Tons)'].str.replace('.','')
df.head()
```

```python
df['Production per Person (Kg)']=df['Production per Person (Kg)'].str.replace('.','')
df['Production per Person (Kg)']=df['Production per Person (Kg)'].str.replace(',','')

df.head()
```

```python
df['Acreage (Hectare)']=df['Acreage (Hectare)'].str.replace('.','')
df['Acreage (Hectare)']=df['Acreage (Hectare)'].str.replace(',','')

df.head()
```

```python
df['Yield (Kg / Hectare)']=df['Yield (Kg / Hectare)'].str.replace('.','')
df['Yield (Kg / Hectare)']=df['Yield (Kg / Hectare)'].str.replace(',','')

df.head()
```

```python
df.rename(
    columns={
        'Production (Tons)' : 'Production_Tons',
        'Production per Person (Kg)' : 'Production_per_Person_kg',
        'Acreage (Hectare)' : 'Acreage_Hectare',
        'Yield_Kg/Hectare' : 'Yield_Kg/Hectare'

    },
    inplace=True
)
df.head()
```

```python
df.isnull().sum()
```

```python
df[df.Acreage_Hectare.isnull()]
```

```python
df.dropna(inplace=True)
df.head(100)
```

```python
df.reset_index(drop=True,inplace=True)
df.head(100)
```

```python
df.columns
```

```python
for i in df.columns[2:]:
    df[i]=df[i].astype('float')
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df.nunique()
```

### How many countries produce sugercane from each continent

```python
cont=df.Continent.value_counts()
cont
```

```python
plt.figure(figsize=(10,6))
plt.title("Top sugercane producing continants")
plt.bar(cont.keys(),cont)
plt.show()
```

### Which countries produces more sugarcane

```python
df.loc[df['Production_Tons'].idxmax(), 'Country']
```

```python
# : → selects all rows
# ["Country", "Production(Tons)"] → selects only these two columns
new_df = df.loc[:, ["Country", "Production_Tons"]].sort_values(by="Production_Tons", ascending=False)
new_df.head()
```

```python
#percentage of production
new_df["Production(Tons)_Percentage"] = (new_df["Production_Tons"]*100)/sum(new_df["Production_Tons"])
new_df.head()
```

```python
top5=new_df.head(5)
plt.figure(figsize=(8, 5))
plt.pie(
    top5['Production(Tons)_Percentage'],
    labels=top5['Country'],
    autopct='%0.1f%%' # -->shows percentage values
)
plt.title('Top 5 Sugar Production Countries (%)')
plt.show()
```
