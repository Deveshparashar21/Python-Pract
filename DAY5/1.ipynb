```python
class Vehicle:
    def __init__(self,Vehicle_id,brand):
        self.Vehicle_id=Vehicle_id
        self.brand=brand
    def calculate_rent(self):
        pass
class Car(Vehicle):
    def __init__(self,Vehicle_id,brand,price_per_day,days):
        super().__init__(Vehicle_id,brand)
        self.price_per_day=price_per_day
        self.days=days
    def calculate_rent(self):
        return self.price_per_day*self.days
class Bike(Vehicle):
    def __init__(self,Vehicle_id,brand,price_per_hour,hour):
        super().__init__(Vehicle_id,brand)
        self.price_per_hour=price_per_hour
        self.hour=hour
    def calculate_rent(self):
        return self.price_per_hour*self.hour
class Truck(Vehicle):
    def __init__(self,Vehicle_id,brand,price_per_km,distance):
        super().__init__(Vehicle_id,brand)
        self.price_per_km=price_per_km
        self.distance=distance
    def calculate_rent(self):
        return self.price_per_km*self.distance
v1=Car("C001","Toyota",1000,5)
v2=Bike("B001","Honda",100,10)
v3=Truck("T001","Volvo",50,200)
print("Car Rent:",v1.calculate_rent())
print("Bike Rent:",v2.calculate_rent())
print("Truck Rent:",v3.calculate_rent())    
```
