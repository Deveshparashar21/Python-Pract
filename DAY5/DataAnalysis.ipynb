#Merge and concat

```python
data1={
    "Name":['A','B','C'],
    "Age":[21,22,23],
    "salary":[1000,2000,3000]
}
data2={
    "Name":['A','BB','C'],
    "Age":[21,27,23],
    "Salary":[1000,2500,3000]
}
```

```python
new_df1=pd.DataFrame(data1)
new_df2=pd.DataFrame(data2)
```

```python
new_df1.head()
```

```python
pd.concat([new_df1,new_df2],ignore_index=True)
```

```python
new_df=pd.concat([new_df1,new_df2],ignore_index=True,axis=0)
```

```python
new_df.drop_duplicates(inplace=True,)
new_df.head()
```

```python
new_df1.head()
```

```python
pd.merge(new_df1,new_df2,on="Age",how="outer")
```
